## Creating a stopwords list

```python
stop_words=["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
            "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
            "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
            "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
            "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
            "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
            "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
            "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
            "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
            "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
```

```python
## Read the data
# phone_reviews = basePath + "datasets\\csv_reviews\\amazon_reviews.csv"
# column = "body"
earphone_reviews = basePath + "datasets\\csv_reviews\\AllProductReviews.csv"
column = "ReviewBody"
data_as_var = pd.read_csv(earphone_reviews)
contents = data_as_var[column]
print(contents[0:10])
```

```python
nlp = spacy.load("en_core_web_sm")

keyword_map = dict()

def keyword_test1(reviews):
    for review in reviews:
        doc = nlp(review)
        for token in doc:
            if token.pos_ == "NOUN" and token.text.lower() not in stop_words:
                if len(token.text) > 1:
                    text = token.text.lower()
                else:
                    text = token.text
                if text in keyword_map:
                    keyword_map[text] = keyword_map[text] + 1
                else:
                    keyword_map[text] = 1
    return keyword_map

keyword_list_sorted = keyword_test1(contents[0:10])
keyword_list_sorted = sorted(keyword_list_sorted.items(),
                             key=lambda item: item[1], reverse=True)
print(keyword_list_sorted[0:30])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

keyword_list = []
for x in keyword_list_sorted:
    keyword_list.append(x[0])

wc = WordCloud(background_color="white",
               colormap="Dark2",
               max_font_size=150,
               random_state=42)

wc.generate(" ".join(keyword_list))

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
